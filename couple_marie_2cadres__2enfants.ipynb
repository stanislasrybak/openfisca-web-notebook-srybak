```python
from __future__ import division

# on fait l'import des differents paquages qui vont nous servir
import collections
import copy
import json
import urllib2
```

```python
# on crée un objet "base_scenario" dans lequel on crée notre ménage
base_scenario = dict(
   #Nombres de personnes dans le foyer et leur rôle
   familles = [
        dict(
            parents = ['ind0', 'ind1'],
            enfants = ['ind2', 'ind3'],
            ),
        ],
    # Qui déclara les revenus et les enfants eventuellements a charge
    foyers_fiscaux = [
        dict(
            declarants = ['ind0', 'ind1'],
            personnes_a_charge = ['ind2', 'ind3'],
            ),
        ],
    # Renseignements sur les individus du foyer
    individus = [
        # Un individu ici Actif, née le 1er Janvier 1970, Cadre, qui a pour ID = "ind0", un salaire de 35000/an et est marié
        dict(
            activite = u'Actif occupé',
            birth = '1970-01-01',
            cadre = True,
            id = 'ind0',
            sali = 35000,
            statmarit = u'Marié',
            ),
        # Un individu ici Actif, née le 2 Janvier 1970, Cadre, qui a pour ID = "ind1", un salaire de 35000/an et est marié
        dict(
            activite = u'Actif occupé',
            birth = '1970-01-02',
            cadre = True,
            id = 'ind1',
            sali = 35000,
            statmarit = u'Marié',
            ),
        # Un individu ici Etudiant, née le 3 Janvier 2000, qui a pour ID = "ind2" et aucun salaire
        dict(
            activite = u'Étudiant, élève',
            birth = '2000-01-03',
            id = 'ind2',
            ),
        # Un individu ici Etudiant, née le 4 Janvier 2000, qui a pour ID = "ind3" et aucun salaire
        dict(
            activite = u'Étudiant, élève',
            birth = '2000-01-04',
            id = 'ind3',
            ),
        ],
    legislation_url = 'http://api.openfisca.fr/api/1/default-legislation',
    menages = [
       # Ici on déclare le ménage
        dict(
            # La personne de référence est le cadre née le 1er Janvier 1970 et qui a pour ID : "ind0"
            personne_de_reference = 'ind0',
            # Son conjoint est le cadre née le 2 Janvier 1970 et qui a pour ID : "ind1"
            conjoint = 'ind1',
            # enfin il y a deux enfants dans le ménage, ils ont pour ID : "ind2" et "ind3"
            enfants = ['ind2', 'ind3'],
            ),
        ],
    )
# La déclaration est faite sur l'année 2013
years = [2013]
```

```python
# on crée un objet "scenarios"
scenarios = []
for year in years:
    # dans cet objet scénario, on y copie les infos rentrés dans notre "base_scenario" dans une variable scenario sous forme de liste grace au ".append"
    scenario = copy.deepcopy(base_scenario)
    scenario['year'] = year
    scenarios.append(scenario)

# on crée une variable simulation dans lequel on met notre scenario (comprend la "base_scenario")
simulation = dict(
    scenarios = scenarios,
    )

# on envoi notre requete pour calculer notre fiscalité et on met le résulat (brut) retourné dans un variable tree
request = urllib2.Request('http://api.openfisca.fr/api/1/simulate', headers = {
    'Content-Type': 'application/json',
    'User-Agent': 'OpenFisca-Notebook',
    })
try:
    # on encode notre réponse  de la simulation en Json grâce au "json.dumps"
    response = urllib2.urlopen(request, json.dumps(simulation))
except urllib2.HTTPError as response:
    print response.read()
    raise
response_dict = json.loads(response.read(), object_pairs_hook = collections.OrderedDict)
trees = response_dict['value']
```

```python
def iter_columns_from_tree(node, base_value = 0, code = None):
    children = node.get('children')
    if children:
        child_base_value = base_value
        for child_code, child in children.iteritems():
            for column in iter_columns_from_tree(child, base_value = child_base_value, code = child_code):
                yield column
            child_base_value += child['values'][value_index]

    value = node['values'][value_index]
    if value != 0 and code is not None:
        column = dict(
            base_value = base_value,
            code = code,
            value = value,
            )
        column.update(node)
        yield column
```

```python
value_index = 0
tree = trees[0]
for column_index, column in enumerate(iter_columns_from_tree(tree)):
    print u'{:20s} | {:10.2f}'.format(column['name'], column['value'])
```
